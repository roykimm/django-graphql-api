```python
class Board(graphene.ObjectType):
    bno = graphene.ID()
    btitle = graphene.String()
    bwriter = graphene.String(required=False)
    bdate = graphene.DateTime(required=False)
    
```

```python
class Query(graphene.ObjectType):
    boards = graphene.List(Board)
    
    def resolve_boards(self, info):
        return [
            Board(btitle='Title1입니다.', bwriter='2', bdate=datetime.now()),
            Board(btitle='Title2입니다.', bwriter='2', bdate=datetime.now()),
            Board(btitle='Title3입니다.', bwriter='2', bdate=datetime.now())
        ]
    
schema = graphene.Schema(query=Query)

result1 = schema.execute(
    '''
    {
        boards {
            btitle
            bdate
        }
    }
    '''
)
```

```python
print(result1)
```

```python
class CreateBoard(graphene.Mutation):
    
    class Arguments:
        btitle = graphene.String()
        bwriter = graphene.String()
        
    board = graphene.Field(Board)
    
    def mutate(self,info, btitle, bwriter):
        
        if info.context.get('is_valid'):
            btitle = btitle.upper()
        
        board = Board(btitle=btitle, bwriter=bwriter)
        return CreateBoard(board=board)

class Mutations(graphene.ObjectType):
    create_board = CreateBoard.Field()
    
schema = graphene.Schema(mutation=Mutations)

result5 = schema.execute(
    '''
    mutation createBoard($btitle: String) {
        createBoard(btitle: $btitle, bwriter:"admin1") {
            board {
                btitle
            }
        }
    }
    ''',
    variable_values = {'btitle': '제목 5번입니다.  im body'},
    context={'is_valid':True}
)

print(result5.data)
```
